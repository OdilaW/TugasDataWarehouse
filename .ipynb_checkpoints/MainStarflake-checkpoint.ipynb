```python
import pandas as pd
import sqlite3
import math
import datetime 
import numpy as np
import matplotlib.pyplot as plt

from IPython.core.display import display, HTML
```

```python
conn = sqlite3.connect('database.db')
```

```python
c = conn.cursor()
```

# DESAIN DATABASE DATAWAREHOUSE
<img src="desaindatabase.png">

```python
c.executescript('drop table if exists Perusahaan')
c.executescript('drop table if exists Konsumen')
c.executescript('drop table if exists Pemesanan')
c.execute("CREATE TABLE IF NOT EXISTS 'Perusahaan' ('Id' INT, 'Nama_perusahaan' VARCHAR(100) NULL, PRIMARY KEY ('Id'))")
c.execute("CREATE TABLE IF NOT EXISTS 'Konsumen' ('Id' INT, 'Username' VARCHAR(100) NULL, 'Alamat'VARCHAR(100) NULL,'no_hp' INT NULL, PRIMARY KEY ('Id'))")
c.execute("CREATE TABLE IF NOT EXISTS 'Pemesanan' ('Id' INT, 'tgl' DATE NULL, 'Jumlah_ayam' INT NULL, 'kg' INT NULL, 'Harga_unit' INT NULL, 'Total_harga' INT NULL, 'Perusahaan_Id' INT NOT NULL, 'Konsumen_Id' INT NOT NULL, PRIMARY KEY ('Id'), FOREIGN KEY ('Perusahaan_Id') REFERENCES 'Perusahaan' ('Id') ON DELETE NO ACTION ON UPDATE NO ACTION, FOREIGN KEY ('Konsumen_Id') REFERENCES 'Konsumen' ('Id') ON DELETE NO ACTION ON UPDATE NO ACTION)")
```

```python
data = pd.read_csv("Data Penjualan Ayam.csv")
data
```

```python
Perusahaan = []
Pemesanan = []
Konsumen  = [] 

for i, j in data.iterrows():
    if j[5] not in Perusahaan:
        Perusahaan.append(j[5])
    
    p = -1
    for index,values in enumerate(Perusahaan):
        if values == j[5]:
            p = index+1
            break
    
    
    if [j[6], j[7], j[8]] not in Konsumen :
        Konsumen.append([j[6], j[7], j[8]])
    
    k = -1
    for index,values in enumerate(Konsumen):
        if values[0] == j[6]:
            k = index+1
            break
    
    if[j[0], j[1], j[2], j[3], j[4], p, k] not in Pemesanan:
        Pemesanan.append([j[0], j[1], j[2], j[3], j[4], p, k])

print(Perusahaan)
print(Pemesanan)
print(Konsumen)
```

```python
c.execute("DELETE FROM PERUSAHAAN")
for index, i in enumerate(Perusahaan): 
    c.execute("INSERT INTO PERUSAHAAN VALUES ({},'{}')".format(index ,i))
    
c.execute("DELETE FROM KONSUMEN")
for index, i in enumerate(Konsumen): 
    c.execute("INSERT INTO Konsumen VALUES ({},'{}','{}','{}')".format(index ,i[0],i[1],i[2]))
    
c.execute("DELETE FROM PEMESANAN")
for index, i in enumerate(Pemesanan): 
    c.execute("INSERT INTO PEMESANAN VALUES ({},'{}','{}','{}','{}','{}','{}','{}')".format(index ,i[0],i[1],i[2],i[3],i[4],i[5],i[6]))
    
conn.commit()
```

```python
c.execute('''SELECT * FROM PERUSAHAAN''')

PERUSAHAAN = []
for row in c.fetchall():
    PERUSAHAAN.append(row)
print("================================================DATA PERUSAHAAN=======================================================")
print(PERUSAHAAN)
print("========================================================================================================================")
print("")
c.execute('''SELECT * FROM KONSUMEN''')

KONSUMEN = []
for row in c.fetchall():
    KONSUMEN.append(row)
print("===============================================DATA KONSUMEN==========================================================")
print(KONSUMEN)
print("========================================================================================================================")
print("")

c.execute('''SELECT * FROM PEMESANAN''')

PEMESANAN = []
for row in c.fetchall():
    PEMESANAN.append(row)
print("================================================DATA PEMESANAN========================================================")
print(PEMESANAN)
print("======================================================================================================================")
print("")

```
